## Imports

```python
import numpy as np
import pandas as pd

from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator

simulator = AerSimulator()
```

## Exercise 1 & 2

### Generate $|00\rangle$

```python
qc = QuantumCircuit(2)

qc.barrier()

# Create a Bell state |00⟩
qc.h(0)
qc.cx(0,1)

qc.measure_all()
qc.draw()
```

```python
# Simulate circuit
job = simulator.run([qc], shots=1024)

# Measure results
counts = job.result().get_counts()

# Show results
print(counts)
#plot_histogram(counts)
```

## Exercise 3

### Select message to send

```python
A = QuantumRegister(1, 'A')
B = QuantumRegister(1, 'B')

MB = ClassicalRegister(2, 'MB')

```

```python
Situation = {}
Situation["Alice_Bit"] = np.random.choice([0, 1])
Situation["Bob_Bit"] = np.random.choice([0, 1])
```

```python
qc = QuantumCircuit(A, B, MB)

# Create a Bell state |00⟩
qc.h(A)
qc.cx(A, B)

qc.barrier()

# Choose message to send
if Situation["Bob_Bit"] == 1:
    qc.z(A)
if Situation["Alice_Bit"] == 1:
    qc.x(A)

qc.barrier()

# Detangle
qc.cx(A, B)
qc.h(A)

qc.barrier()

# Bob measures both qubits
qc.measure(A, MB[0])
qc.measure(B, MB[1])
qc.draw()
```

```python
# Simulate circuit
job = simulator.run([qc], shots=1024)

# Measure results
counts = job.result().get_counts()

# Show results
print(counts)
#plot_histogram(counts)
```

```python
print(f"Chosen message: {Situation['Alice_Bit']}{Situation['Bob_Bit']} | Received message: {list(counts.keys())[0]}")
```
